```python
import matplotlib.pyplot as plt # Import needed modules here
```

```python
def StateData(state,caseDeath): 
    caseData = open("/home/hh11123/myBiologyClasses/Python/4.16_week11/covidData_4.7.20.csv","r") #get data from the file
    caseDictionary = {} #create two dictionaries to hold the data and the number
    deathDictionary = {}

    for i in caseData:
        if state in i: #get the data from the state we want
            date, county, state, fips, cases, deaths=i.split(",") #split the str by ","
            if caseDeath=="cases":#check the user want to know case or death statistic
                case = int(cases) #cast the cases to integer
                if date in caseDictionary:
                    caseDictionary[date] = caseDictionary[date] + case# check the date is thereor not, if it is, add more cases in dictionary
                else:
                    caseDictionary[date] = case
                
            elif caseDeath=="deaths":
                death =int(deaths)
                if date in deathDictionary:
                    deathDictionary[date]= deathDictionary[date] + death
                else:
                    deathDictionary[date] = death

    caseData.close()
    
    x_axle =[]
    y_axle =[]
    x_axle = list(caseDictionary.keys())
    y_axle = list(caseDictionary.values())

    return x_axle,y_axle
```

```python
stateInput = input("Enter a state you want: ")
case_deathInput = input("Enter whether case counts or deaths you want: ")
xaxis, yaxis = StateData(state = stateInput, caseDeath = case_deathInput)#get the result from def and give the value into x and y axis
plt.figure()
plt.xlabel('Date')
plt.ylabel('The number of {} in {}'.format(stateInput, case_deathInput))

plt.xticks(rotation=90)
plt.grid(True)# generate grids
plt.plot(xaxis,yaxis)
plt.show()
```

```python
def countyData(county,state,caseDeath):
    caseData = open("/home/hh11123/myBiologyClasses/Python/4.16_week11/covidData_4.7.20.csv","r")
    caseDictionary = {}
    deathDictionary = {}

    for i in caseData:
        if county in i and state in i:# get the data from the exactly county and state we want
            date, county, state, fips, cases, deaths=i.split(",")
            if caseDeath=="cases":
                case = int(cases)
                if date in caseDictionary:
                    caseDictionary[date] = caseDictionary[date] + case
                else:
                    caseDictionary[date] = case
                
            elif caseDeath=="deaths":
                death =int(deaths)
                if date in deathDictionary:
                    deathDictionary[date]= deathDictionary[date] + death
                else:
                    deathDictionary[date] = death

    caseData.close()
    
    x_axle =[]
    y_axle =[]
    x_axle = list(caseDictionary.keys())
    y_axle = list(caseDictionary.values())

    return x_axle,y_axle
```

```python
countyInput = input("Enter a county you want: ")
stateInput = input("Enter a state you want: ")
case_deathInput = input("Enter whether case counts or deaths you want: ")
xaxis, yaxis = countyData(county = countyInput, state = stateInput, caseDeath = case_deathInput)#get the result from the def and put the value into x and y axis
plt.figure()
plt.xlabel('Date')
plt.ylabel('The number of {} in {} , {}'.format(case_deathInput, countyInput, stateInput)) 
#automaticlly generating a title based on the county and state

plt.xticks(rotation=90)
plt.grid(True)
plt.plot(xaxis,yaxis)
plt.show()
```
